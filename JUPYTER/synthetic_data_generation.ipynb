```python
# Importing the necessary libraries to generate synthetic data for the actors table. 
# The pandas library is used to create a dataframe and save it as a csv file. 
# The faker library is used to generate random names, birthdates, and other details for the actors.

import pandas as pd
import random
from faker import Faker

 # Setting the locale to English (Ireland) for generating names and other details.

fake = Faker('en_IE')

users=[]

for i in range(1000):

    users.append({

        "userName": fake.user_name(),

        "firstName": fake.first_name(),

        "surName": fake.last_name(),

        "password": fake.password(length=15)

    })

df_users = pd.DataFrame(users)

df_users.to_csv(
    "../DATASETS/users.csv",
    index=False
)

# Displaying the first few rows of the dataframe and the total number of users generated.
print(df_users.head())
print("Total users:",len(df_users))
```

```python
# Creating a list of dictionaries representing famous actors.
famous_actors = [
    {"firstName": "Meryl", "surName": "Streep", "nationality": "American"},
    {"firstName": "Denzel", "surName": "Washington", "nationality": "American"},
    {"firstName": "Jack", "surName": "Nicholson", "nationality": "American"},
    {"firstName": "Al", "surName": "Pacino", "nationality": "American"},
    {"firstName": "Morgan", "surName": "Freeman", "nationality": "American"},
    {"firstName": "Christian", "surName": "Bale", "nationality": "British"},
    {"firstName": "Emma", "surName": "Watson", "nationality": "British"},
    {"firstName": "Benedict", "surName": "Cumberbatch", "nationality": "British"},
    {"firstName": "Gary", "surName": "Oldman", "nationality": "British"},
    {"firstName": "Tilda", "surName": "Swinton", "nationality": "British"},
    {"firstName": "Saoirse", "surName": "Ronan", "nationality": "Irish"},
    {"firstName": "Evanna", "surName": "Lynch", "nationality": "Irish"},
    {"firstName": "Brendan", "surName": "Gleeson", "nationality": "Irish"},
    {"firstName": "Saoirse", "surName": "Ronan", "nationality": "Irish"},
    {"firstName": "Ryan", "surName": "Gosling", "nationality": "Canadian"},
    {"firstName": "Margot", "surName": "Robbie", "nationality": "Australian"},
    {"firstName": "Cate", "surName": "Blanchett", "nationality": "Australian"},
    {"firstName": "Timothée", "surName": "Chalamet", "nationality": "French"},
    {"firstName": "Marion", "surName": "Cotillard", "nationality": "French"},
    {"firstName": "Javier", "surName": "Bardem", "nationality": "Spanish"},
]

print("=" * 70)
print("ACTORS TABLE - Famous Actors")
print("=" * 70)
print("\nLista słynnych aktorów (20):")
for i, actor in enumerate(famous_actors, 1):
    print(f"{i:2d}. {actor['firstName']:15} {actor['surName']:20} ({actor['nationality']})")

# Generating a list of actors by combining the famous actors with randomly generated ones using the Faker library.
# The faker library is used to generate random names and nationalities for the actors. 
# The generated data is stored in a pandas dataframe and saved as a csv file.
actors = []

# Adding the famous actors to the list of actors.
for actor in famous_actors:
    actors.append({
        "actorFirstName": actor["firstName"],
        "actorSurName": actor["surName"],
        "nationality": actor["nationality"]
    })

# Calculating the number of actors to generate randomly to reach a total of 100 actors, and generating the random actors using the Faker library.
num_generated = 100 - len(famous_actors)
for i in range(num_generated):
    actors.append({
        "actorFirstName": fake.first_name(),
        "actorSurName": fake.unique.last_name(),
        "nationality": fake.country()
    })

df_actors = pd.DataFrame(actors)
df_actors.to_csv("../DATASETS/actors.csv", index=False)

print(df_actors.head(10))
```

```python
# Generating synthetic data for the directors table. 
# The faker library is used to generate random names and nationalities for the directors.

directors=[]

# Generate 100 directors
for i in range(100):

    directors.append({

        "directorFirstName": fake.first_name(),

        "directorSurName": fake.unique.last_name(),

        "nationality": fake.country()

    })

df_directors = pd.DataFrame(directors)

df_directors.to_csv(
    "../DATASETS/directors.csv",
    index=False
)

print(df_directors.head())
```

```python
# Generating synthetic data for the movies table. 
# The movies will be generated in two parts:
# 1. Movies with famous actors (70% of the total), where the same famous actors will appear in multiple movies.
# 2. Movies with randomly generated actors (30% of the total), where the actors will be unique and generated using the Faker library.   

genres = [
    "Action",
    "Drama",
    "Comedy",
    "Horror",
    "Sci-Fi",
    "Romance",
    "Thriller"
]

movies = []

# Defining the number of movies to generate for each category (famous actors and other actors).
num_famous_movies = 800  
num_other_movies = 300   


# Generating movies with famous actors. The same famous actors will appear in multiple movies, and the details of the movies (name, genre, release year, duration, rating) will be generated randomly using the Faker library and the random module.

for i in range(num_famous_movies):
    # Randomly selecting a famous actor for the movie.
    actor = random.choice(famous_actors)
    
    movies.append({
        "movieName": fake.sentence(nb_words=3),
        "genre": random.choice(genres),
        "releaseYear": random.randint(1990, 2026),
        "duration": random.randint(60, 180),
        "rating": round(random.uniform(1, 5), 1),
        "actorFirstName": actor["firstName"],
        "actorSurName": actor["surName"],
        "actor_id": famous_actors.index(actor) + 1,  # ID: 1-20 (słowni)
        "director_id": random.randint(1, 100)
    })

# Generating movies with randomly generated actors. Each movie will have a unique actor generated using the Faker library, and the details of the movies will be generated randomly as well.

for i in range(num_other_movies):
    movies.append({
        "movieName": fake.sentence(nb_words=3),
        "genre": random.choice(genres),
        "releaseYear": random.randint(1990, 2026),
        "duration": random.randint(60, 180),
        "rating": round(random.uniform(1, 5), 1),
        "actorFirstName": fake.first_name(),
        "actorSurName": fake.unique.last_name(),
        "actor_id": random.randint(21, 100),  # ID: 21-100 (zwykli)
        "director_id": random.randint(1, 100)
    })

df_movies = pd.DataFrame(movies)
df_movies.to_csv("../DATASETS/movies.csv", index=False)


print(df_movies.head())
```

```python
# Generating synthetic data for the series table. 
# The faker library is used to generate random names for the series, actors, and directors.

genres = [
    "Action",
    "Drama",
    "Comedy",
    "Horror",
    "Sci-Fi",
    "Romance",
    "Thriller"
]

series=[]

# Generate 1000 series
for i in range(1000):

    series.append({

        "seriesName": fake.sentence(nb_words=3),

        "genre": random.choice(genres),

        "releaseYear": random.randint(1990,2026),

        "actorFirstName": fake.first_name(),

        "actorSurName": fake.unique.last_name(),

        "actor_id": random.randint(1,100),

        "rating": round(
            random.uniform(1,5),
            1
        ),

        "numberOfSeasons": random.randint(1,15),

        "director_id": random.randint(1,100)

    })

df_series = pd.DataFrame(series)

df_series.to_csv(
    "../DATASETS/series.csv",
    index=False
)

print(df_series.head())
```

```python
# Generating synthetic data for the user_movies table. 
# The faker library is used to generate random user IDs and movie IDs, and the genres are randomly selected from a predefined list of genres.


genres = [
    "Action",
    "Drama",
    "Comedy",
    "Horror",
    "Sci-Fi",
    "Romance",
    "Thriller"
]

user_movies=[]

# Generate 1000 viewing activities
for i in range(1000):

    user_movies.append({

        "user_id": random.randint(1,1000),

        "movie_id": random.randint(1,1000),

        "genre": random.choice(genres)

    })

df_user_movies = pd.DataFrame(user_movies)

df_user_movies.to_csv(
    "../DATASETS/user_movies.csv",
    index=False
)

print(df_user_movies.head())
```

```python
# Generating synthetic data for the user_series table. 
# The faker library is used to generate random user IDs and series IDs, and the genres are randomly selected from a predefined list of genres.

genres = [
    "Action",
    "Drama",
    "Comedy",
    "Horror",
    "Sci-Fi",
    "Romance",
    "Thriller"
]

user_series=[]

# Generate 1000 user activities
for i in range(1000):

    user_series.append({

        "user_id": random.randint(1,1000),

        "series_id": random.randint(1,1000),

        "genre": random.choice(genres)

    })

df_user_series = pd.DataFrame(user_series)

df_user_series.to_csv(
    "../DATASETS/user_series.csv",
    index=False
)

print(df_user_series.head())
```
