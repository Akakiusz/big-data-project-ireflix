### ***Solution:
Our MapReduce solution processes Ireflix viewing logs to analyze genre 
popularity based on real database data from CA1. The implementation 
processes sample (from movies.csv) viewing records covering 6 different genres: Action, 
Drama, Comedy, Horror, Sci-Fi, and Thriller.

```python
# Importing necessary libraries

from collections import defaultdict
import pandas as pd
```

 Using synthetic movie dataset generated for Ireflix database

```python
# INPUT DATA: Load CSV

data = pd.read_csv('../DATASETS/movies.csv')
```

```python
# MAPPER FUNCTION 

def map_function(data):
    mapped = []
    for genre in data ['genre']:
        mapped.append((genre, 1))   
    return mapped
```

```python
# SHUFFLE AND SORT.

def shuffle_sort(mapped_data):
    grouped_data = defaultdict(list)
    for key, value in mapped_data:
        grouped_data[key].append(value)
    return grouped_data
  
```

```python
# REDUCER FUNCTION.

def reduce_function(grouped_data):
    reduced_data = {}
    for key, values in grouped_data.items():
        reduced_data[key] = sum(values)
    return reduced_data
```

```python
# EXECUTE MAPREDUCE

mapped_data = map_function(data)
grouped_data = shuffle_sort(mapped_data)
reduced_data = reduce_function(grouped_data)
print(reduced_data)
print(mapped_data)
print(grouped_data)
```

```python
print("== IREFLIX GENRES ANALYSIS ==")

for genre, total in reduced_data.items():
    print(genre, ":", total)
```

The MapReduce analysis shows that Drama (165) and Horror(164) are among the most watched genres (for movies) on the Ireflix platform.
